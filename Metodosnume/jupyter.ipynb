```python
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 3*x**2 + x - 1

def df(x):
    return 3*x**2 - 6*x + 1

x_vals1 = []
x = -10

for i in range(60):
    x_vals1.append(x)
    x = x - f(x)/df(x)

iters1 = list(range(len(x_vals1)))


x_vals2 = []
x = -8

for i in range(60):
    x_vals2.append(x)
    x = x - f(x)/df(x)

iters2 = list(range(len(x_vals2)))

plt.figure(figsize=(18,6))

# Primera curva
plt.plot(iters1, x_vals1, marker='o', label='x0 = -10')

# Segunda curva
plt.plot(iters2, x_vals2, marker='s', label='x0 = -8')

# Etiquetas de puntos
for i, x in enumerate(x_vals1):
    plt.text(i, x, f"{x:.2f}", fontsize=7)

for i, x in enumerate(x_vals2):
    plt.text(i, x, f"{x:.2f}", fontsize=7)

plt.title("Método de Newton-Raphson")
plt.xlabel("Iteración")
plt.ylabel("xₙ")

plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

# ======================================
# FUNCIÓN Y DERIVADA
# ======================================
def f(x):
    return x**3 - 3*x**2 + x - 1

def df(x):
    return 3*x**2 - 6*x + 1


x_vals = []

x = -10

for i in range(8):
    x_vals.append(x)
    x = x - f(x)/df(x)

iteraciones = np.arange(len(x_vals))

# suavizar curva
x_smooth = np.linspace(iteraciones.min(),
                       iteraciones.max(),
                       300)

spl = make_interp_spline(iteraciones, x_vals, k=3)

y_smooth = spl(x_smooth)

plt.figure(figsize=(14,7))

# curva suave
plt.plot(
    x_smooth,
    y_smooth,
    linewidth=3,
    color='royalblue',
    label='Curva Newton'
)

# puntos reales
plt.plot(
    iteraciones,
    x_vals,
    'o',
    markersize=8,
    color='darkblue',
    label='Iteraciones Newton'
)

# líneas verticales
for i, y in enumerate(x_vals):

    plt.vlines(
        i,
        0,
        y,
        colors='gray',
        linestyles='dashed',
        alpha=0.7
    )

    # texto
    plt.text(
        i,
        y,
        f"{y:.2f}",
        fontsize=9
    )


# eje X
plt.axhline(0, color='black', linewidth=1)

# diseño
plt.title(
    "Convergencia del Método de Newton-Raphson",
    fontsize=18
)

plt.xlabel("Iteración", fontsize=14)
plt.ylabel("xₙ", fontsize=14)

plt.grid(True, alpha=0.3)

plt.legend(fontsize=12)

plt.show()
```
